```python
import numpy as np
import pandas as pd
import warnings 
warnings.filterwarnings(action='ignore')
```

```python
df=pd.read_csv("SMSSpamCollection.txt",sep='\t', names=['label','message'])
```

```python
import  re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
corpus=[]
ps=PorterStemmer()
for i in range(len(df)):
    temp=re.sub('[^a-zA-Z]',' ',df['message'][i])
    temp=temp.lower()
    temp=temp.split()
    temp=[ps.stem(word)  for word in temp if word not in stopwords.words('english')]
    temp=" ".join(temp)
    corpus.append(temp)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,binary=True,ngram_range=(1,3))
x=cv.fit_transform(corpus).toarray()
```

```python
x
```

```python
cv.vocabulary_
```

```python
y=pd.get_dummies(df['label'])
y
```

```python
y=y.iloc[:,0].values
```

```python
y
```

```python
y=y.astype('int')
```

```python
y=pd.DataFrame(y)
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
model=nb.fit(x_train,y_train)
model
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
